```python
def add_day(courses: list[list[Clazz]]):
    """
    :param courses: 课程数量
    """
    for n, i in enumerate(courses):
        for x in i:
            x.day = n + 1


def random_color():
    colors1 = [
        "#364f6b",
        "#112d4e",
        "#48466d",
        "#3d84a8",
        "#88304e",
        "#444f5a",
        "#0c056d",
        "#574b90",
        "#303a52",
        "#9e579d",
        "#3a4750",
        "#4d4545",
        "#ffbd39",
        "#e61c5d",
        "#3a0088",
        "#930077",
        "#071a52",
    ]
    return str(random.choice(colors1)).upper()


def add_cur(days: list[list[Course]]):
    # 不重复的课程
    # 一个班的所有课程
    courses = []
    for d in days:
        for d2 in d:
            courses.append(d2)

    ids = {}
    # 添加id

    for n, i in enumerate(list(set([i.course_name for i in courses]))):
        ids[i] = n
    for i in courses:
        i.id = ids[i.course_name]
    courses_str = []
    for i in ids.items():
        courses_str.append(
            '{' + f'"color":"{random_color()}","courseName":"{i[0]}","credit":0.0,"id":{i[1]},"note":"","tableId":4' + '}')
    courses_info_str = []
    for i in courses:
        courses_info_str.append(
            '{' + f'"day":{i.day},"endTime":"","endWeek":{i.endWeek},"id":{i.id},"level":0,"ownTime":false,"room":"{i.room}","startNode":{i.start_node},"startTime":"","startWeek":{i.start_week},"step":{i.step + 1},"tableId":4,"teacher":"{i.teacher}","type":0' + '}')

    return [
        '[' + ','.join(courses_str) + ']',
        '[' + ','.join(set(courses_info_str)) + ']'
    ]


for i, v in classes.items():
    add_day(v.days)
    with open(f'./2024-2/{i}.wakeup_schedule', 'w', encoding='utf-8') as f:
        f.write(head(i))
        for i in add_cur(v.days):
            f.write('\n' + i)
```
